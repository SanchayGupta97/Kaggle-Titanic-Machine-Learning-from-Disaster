```python
plt.figure(figsize=(10, 10))

for column_index, column in enumerate(titanic_train.columns):
    if column == 'Pclass':
        continue
    plt.subplot(3, 3, column_index + 1)
    sb.violinplot(x='Pclass', y=column, data=titanic_train)
```

```python
plt.figure(figsize=(10, 10))
for column_index, column in enumerate(titanic_train.columns):
    if column == 'Pclass':
        continue
    plt.subplot(3, 3, column_index + 1)
    sb.boxplot(x='Pclass', y=column, data=titanic_train)
```
